## **Dataset Overview: Breast Cancer Dataset**

The **Breast Cancer** dataset is a well-known dataset available in the sklearn library, primarily used for classification. The dataset contains information about various features of cell nuclei from breast cancer biopsies, with the goal of predicting whether the tumor is **malignant** (cancerous) or **benign** (non-cancerous).

The dataset contains 30 features, derived from digital images of fine needle aspirate (FNA) of breast cancer tumors. These features represent various measurements related to the shape and texture of the cell nuclei.

## **Data Acquisition**

```python
# importing the necessary libraries
import pandas as pd
from sklearn.datasets import load_breast_cancer

# loading the Breast Cancer dataset from sklearn
data = load_breast_cancer()
```

```python
# converting the dataset to a pandas DataFrame
df = pd.DataFrame(data.data, columns=data.feature_names)

# adding the target column to the DataFrame
df['target'] = data.target
```

```python
# displaying the first 5 rows of the dataset
df.head()
```

```python
df.info()
```

Clearly, there are no missing values in the dataset.

```python
# checking for duplicates
print(f"Duplicate rows : {df.duplicated().sum()}")
```

```python
# Statistical analysis of the dataset
print("Descriptive Statistics:\n")
df.describe()
```

**Observation :** The dataset's datatypes are correct and has no duplicate rows, hence it is clean and we can move to data preprocessing steps.

## **Data Visualization and Feature Selection**

```python
import matplotlib.pyplot as plt

# plotting the distribution
class_counts = df['target'].value_counts()
plt.figure(figsize=(8, 6))
class_counts.plot(kind='bar', color=['red', 'blue'], alpha=0.7)
plt.title("Class Distribution (Malignant vs Benign)", fontsize=14)
plt.xticks(ticks=[0, 1], labels=['Benign', 'Malignant'], rotation=0)
plt.ylabel("Count")
plt.xlabel("Class")
plt.show()
```

The dataset depicts slight class imbalance as there are 357 and 212 records belonging to benign and malignant class.

```python
import seaborn as sns

# computing correlations
correlations = df.iloc[:, :-1].corr()

# plotting the heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(correlations, cmap='coolwarm', annot=False, fmt=".2f")
plt.title("Feature Correlation Heatmap", fontsize=16)
plt.show()
```

After analyzing the correlation matrix and eliminating features having correlation greater than 0.75, the following features were selected to avoid redundancy and improve model performance (robustness):
1. Mean Area
2. Mean Texture
3. Area Error
4. Mean Smoothness
5. Mean Concave Points
6. Worst Fractal Dimension
7. Worst Compactness
8. Worst Concave Points

```python
top_features = ['mean area', 'mean texture', 'area error', 'mean smoothness', 
                'mean concave points', 'worst fractal dimension', 
                'worst compactness', 'worst concave points']

# calculating the number of rows and columns for subplots
n_features = len(top_features)
n_cols = 4  # Number of columns
n_rows = (n_features + n_cols - 1) // n_cols

# creating subplots
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(16, 12), sharey=False)

# flattening axes for easy iteration
axes = axes.flatten()

# plotting each feature
for i, feature in enumerate(top_features):
    sns.boxplot(
        data=df, 
        x='target', 
        y=feature, 
        ax=axes[i], 
        hue='target',  
        palette={0: 'blue', 1: 'red'},  
        dodge=False, 
        legend=False
    )
    axes[i].set_title(f"{feature}", fontsize=10)
    axes[i].set_xticks([0, 1])
    axes[i].set_xticklabels(['Benign', 'Malignant'])
    axes[i].set_xlabel("Class")
    if i % n_cols == 0:  
        axes[i].set_ylabel("Value")

# removing empty subplots (if any)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# adding the overall title
plt.suptitle("Boxplots of Top Predictive Features by Class", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# standardizing the important features
featuresdf = df[['mean area', 'mean texture', 'area error', 'mean smoothness', 'mean concave points', 
                                         'worst fractal dimension', 'worst compactness', 'worst concave points']]

scaler = StandardScaler()
scaled_data = scaler.fit_transform(featuresdf.iloc[:, :-1])

# implementing PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)
df_pca = pd.DataFrame(pca_result, columns=['PC1', 'PC2'])
df_pca['target'] = df['target']

# plotting the result
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='target', palette={0: 'blue', 1: 'red'}, alpha=0.7)
plt.title("PCA Visualization (2 Components)", fontsize=14)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

```python
# separating features (X) from the target variable (y)
X = df.drop(columns=['target'])
y = df['target']
```

```python
from sklearn.ensemble import RandomForestClassifier

# performing feature selection using RFE with a RandomForest classifier as the estimator
estimator = RandomForestClassifier(random_state=42)
estimator.fit(X, y)

# getting feature importances
feature_importances = estimator.feature_importances_

# creating a DataFrame with feature names and their importance
importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importances
})

# sorting the features by importance
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# selecting top 15 features
top_15_features = importance_df.head(15)

# plotting feature importances as a bar plot
plt.figure(figsize=(10, 6))
plt.barh(top_15_features['Feature'], top_15_features['Importance'], color='skyblue')
plt.xlabel('Importance')
plt.title('Feature Importance (Top 15 Selected Features)', fontsize=16)
plt.gca().invert_yaxis()             # to display the most important features on top
plt.show()
```

## **Data Preprocessing**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_selected = df[['mean area', 'worst radius', 'worst perimeter', 'mean perimeter', 'mean concavity', 'worst area', 'mean concave points',
                 'worst compactness', 'worst concave points', 'mean radius']]

# splitting the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# applying feature scaling to the selected features
scaler = StandardScaler()

# fit and transform on the training data
X_train_scaled = scaler.fit_transform(X_train)  

# transform the test data using the same scaler
X_test_scaled = scaler.transform(X_test) 
```

## **ANN Model Building**

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report

# initializing the Artificial Neural Network (ANN) model with specified parameters
ann = MLPClassifier(random_state=42, max_iter=1000)

# training the ANN model using the scaled training data
ann.fit(X_train_scaled, y_train)

# making predictions on the test set
y_pred = ann.predict(X_test_scaled)

# evaluating and displaying the model's performance
print("ANN Model Evaluation:|n")
print(f"Accuracy: {accuracy_score(y_test, y_pred)}\n") 
print("Classification Report:")
print(classification_report(y_test, y_pred))             # detailed classification metrics (precision, recall, F1-score)
```

## **Hyperparameter Tuning**

```python
from sklearn.model_selection import GridSearchCV

# defining a grid of hyperparameters to tune for the ANN model
param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (150,), (200,)],                    # Number of neurons in the hidden layer
    'activation': ['tanh', 'relu'],                                           # Activation functions to try
    'solver': ['adam', 'sgd'],                                                # Optimization algorithms to test
    'alpha': [0.0001, 0.001, 0.01, 0.1],                                      # Regularization term to avoid overfitting
    'learning_rate': ['constant', 'invscaling', 'adaptive']                   # Learning rate schedules
}

# performing hyperparameter tuning using GridSearchCV with cross-validation
grid_search = GridSearchCV(MLPClassifier(max_iter=500, early_stopping=True, random_state=42), param_grid, cv=5, n_jobs=-1)

# fitting the grid search model to the training data
grid_search.fit(X_train_scaled, y_train)

# printing the best parameters and the best cross-validation score from GridSearchCV
print("Best Parameters Found by GridSearchCV:\n")
print(grid_search.best_params_)
print(f"Best Cross-validation Score: {grid_search.best_score_}\n")

# evaluating the best model found by GridSearchCV on the test set
y_pred_grid = grid_search.best_estimator_.predict(X_test_scaled)
print("GridSearchCV Best Model Evaluation:\n")
print(f"Accuracy: {accuracy_score(y_test, y_pred_grid)}\n")
print("Classification Report:")
print(classification_report(y_test, y_pred_grid))
```

## Model Comparison: ANN vs. ANN with Hyperparameter Tuning

### 1. **ANN Model (Without Hyperparameter Tuning)**

### 2. **ANN Model (With Hyperparameter Tuning)**
- **Accuracy**: 94.74%
- **Precision (Class 0)**: 0.95 | **Recall (Class 0)**: 0.91 | **F1-score (Class 0)**: 0.93
- **Precision (Class 1)**: 0.95 | **Recall (Class 1)**: 0.97 | **F1-score (Class 1)**: 0.96
- The tuned model shows a slightly lower accuracy, precision, recall and f1-score.

### **Conclusion:**
- The **ANN Model (Without Hyperparameter Tuning)** performs better overall with a higher accuracy (99.12% vs. 94.74%).
- The untuned model with robust feature selection is the more effective choice for this dataset.
